## Splitten in Training-, Test- und Validierungs-Datensatz

Dieser Code wurde separat für alle Dokument-Splits ausgeführt, um sicherzustellen dass das Mengenverhältniss der Fragen zu den einzelnen Betriebsanleitungen in den finalen Train-, Validate- und Test-Splits jenem aus dem gesamten Datensatz entspricht.

```python
import json
import random

INPUT_FILE = "datasets/GerManualDPR/manual_doc_splits/full_splits/ctrlX_Core_Split.json"
TRAIN_OUT = "datasets/GerManualDPR/manual_doc_splits/train_splits/train_ctrlX_Core_Split.json"
VALIDATE_OUT = "datasets/GerManualDPR/manual_doc_splits/validate_splits/validate_Core_Drive_Split.json"
TEST_OUT = "datasets/GerManualDPR/manual_doc_splits/test_splits/test_ctrlX_Core_Split.json"

# Verhältnis bestimmen (80% Train, 10% Validate, 10% Test)
TRAIN_RATIO = 0.8
VALIDATE_RATIO = 0.1
TEST_RATIO = 0.1

# Seed für Reproduzierbarkeit
SEED = 314
random.seed(SEED)


# JSON laden
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

# Durchmischen (deterministisch)
random.shuffle(data)

# Split erstellen
n = len(data)
train_end = int(n * TRAIN_RATIO)
validate_end = train_end + int(n * VALIDATE_RATIO)

train_data = data[:train_end]
validate_data = data[train_end:validate_end]
test_data = data[validate_end:]

print(f"Gesamt: {n}")
print(f"Train: {len(train_data)}  ({TRAIN_RATIO*100:.0f}%)")
print(f"Validate: {len(validate_data)}  ({VALIDATE_RATIO*100:.0f}%)")
print(f"Test: {len(test_data)}  ({TEST_RATIO*100:.0f}%)")

# JSON-Dateien speichern
with open(TRAIN_OUT, "w", encoding="utf-8") as f:
    json.dump(train_data, f, indent=2, ensure_ascii=False)

with open(VALIDATE_OUT, "w", encoding="utf-8") as f:
    json.dump(validate_data, f, indent=2, ensure_ascii=False)

with open(TEST_OUT, "w", encoding="utf-8") as f:
    json.dump(test_data, f, indent=2, ensure_ascii=False)

print("✓ Train-, Validate- und Test-Splits gespeichert.")
```

## Mischen des Zusammengesetzten Trainings-Splits
Nachdem die einzelnen Splits der unterschiedlichen Betriebsanleitungen manuell zusammengeführt wurden, werden die Splits nochmals zufällig umsortiert.

```python
INPUT_FILE = "datasets/GerManualDPR/manual_unshuffled/test_dataset_unshuffled.json"
OUTPUT_FILE = "datasets/GerManualDPR/manual/test_dataset.json"

# Seed für Reproduzierbarkeit
SEED = 314
random.seed(SEED)

# JSON laden
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

# Durchmischen (deterministisch)
random.shuffle(data)

# Neue JSON speichern
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print(f"✓ JSON wurde deterministisch geshuffelt und gespeichert als:\n{OUTPUT_FILE}")
```
